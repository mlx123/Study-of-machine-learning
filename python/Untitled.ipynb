```python
import pandas as pd
```

```python
pic = pd.read_csv("./pic.csv",sep = '\ ',names = ['time','size'])#读取csv文件，并指定分隔符和列标签
```

```python
pic
```

```python
#将两列的时间做减法，构成时间差
timsub = pd.to_datetime(pic['time'])-pd.to_datetime(pic.shift(1)['time'])
t2 = pd.DataFrame({'time':timsub}).dropna()#去掉Nan
t3 = []
for i in t2['time']:
    k=0
    t3.append(int(i.seconds))
    k=k+1

t3
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

plt.hist(t3,alpha = 0.5,bins=10,histtype = 'stepfilled',range=[1,3] )
```
